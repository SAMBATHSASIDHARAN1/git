## Importing Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Assumptions
* Assume Salary Rs 10,00,000/-
* Tax Rate 30%
* Annual Inflation 2.5%
* Expenses
    * Monthly Rent 10,0000/- per month
    * Daily Food Budge 200/- day(30 days per month)
    * Entertainment Expense 2000/- month
    * Unforeseen Expenses 3000/- month
* Annual Salary Growth Rate 5%    

```python
salary = 1000000
tax_rate=0.3
monthly_salary_after_tax=salary*(1-tax_rate)/12
monthly_rent=10000
monthly_food_expense=200*30
unforeseen_expense=3000
entertainment_expense=2000
monthly_expense=monthly_rent+monthly_food_expense+entertainment_expense+unforeseen_expense
monthly_saving=salary-monthly_expense
```

```python
forecast_months =6*12  #6 Years
annual_salary_growth_rate=0.05
```

```python
annual_inflation=0.025
monthly_inflation=np.power((1+annual_inflation),1/12)-1
```

```python
cumulative_inflation_forecast=np.cumproduct(np.repeat(1+monthly_inflation,forecast_months))
expense_forecast=monthly_expense*cumulative_inflation_forecast
```

## Plotting Expense

```python
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.plot(expense_forecast,color='black')
plt.title('Expense Forecast')
plt.xlabel('Period')
plt.ylabel('Monthly Expense')
plt.show()
```
